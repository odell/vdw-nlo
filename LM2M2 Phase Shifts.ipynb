```python
import numpy as np
import matplotlib.pyplot as plt

import lm2m2 as he4

gr = 4/3
h = 3
```

```python
R = he4.BETA6/5
```

```python
system = he4.LM2M2System(R, 0)
```

```python
system.bound_state_spectrum(0, 0)
```

```python
momenta = np.linspace(0.01/he4.BETA6, 30/he4.BETA6, 60)
```

```python
phase_shifts = np.arctan(momenta/system.kcotd_gen_fast(momenta, 0, 0))
```

```python
fig, ax = plt.subplots(dpi=300)
fig.patch.set_facecolor('white')

ax.plot(momenta, phase_shifts)

ax.set_xlabel(r'$k$')
ax.set_ylabel(r'$\delta_{\ell=0}$')
ax.set_title(r'LM2M2');
```

```python
np.savetxt('datfiles/lm2m2_swave_phase_shifts.txt', np.vstack((momenta, phase_shifts)).T)
```
