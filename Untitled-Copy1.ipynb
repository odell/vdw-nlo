```python
import numpy as np
import matplotlib.pyplot as plt
import helium4plusplus as he4

plt.style.use('science')
```

```python
rgflow = np.loadtxt('datfiles/rg-flow-semilocal-1.dat')
```

```python
fig, ax = plt.subplots(dpi=300)
fig.patch.set_facecolor('white')

ax.plot(he4.BETA6/rgflow[:, 0], rgflow[:, 1], marker='.')
ax.plot(he4.BETA6/rgflow[:, 0], rgflow[:, 2], marker='.')

ax.set_xlabel(r'$\beta_6/R$')
ax.set_ylabel(r'$g_{\rm LO}$');
# ax.set_yscale('log');
```

```python
def sort_by_cutoff(d):
    ii = np.argsort(d[:, 0])
    return d[ii, :]
```

```python
# Arnas plots
b3_lo = sort_by_cutoff(np.loadtxt('datfiles/deltuva_nonlocal_lo_bgs+.txt'))
b3_nlo = sort_by_cutoff(np.loadtxt('datfiles/deltuva_nonlocal_nlo_bgs+.txt'))

b3_nlo_new = sort_by_cutoff(np.loadtxt('datfiles/x_full'))
b3_nlo_new_short = sort_by_cutoff(np.loadtxt('datfiles/x_short'))
```

```python
fig, ax = plt.subplots(dpi=300)
fig.patch.set_facecolor('white')

ax.plot(b3_lo[:, 0], b3_lo[:, 1], linestyle='-', label=r'$R/10$ LO')
ax.plot(b3_nlo[:, 0], b3_nlo[:, 1], linestyle='--', label=r'$R/10$ NLO')

ax.plot(b3_nlo_new[:, 0], b3_nlo_new[:, 1], linestyle='-.', label=r'$R$ NLO')
ax.plot(b3_nlo_new_short[:, 0], b3_nlo_new_short[:, 1], linestyle=':', label=r'$R$ NLO')

ax.legend()
ax.set_xlabel(r'$R$')
ax.set_ylabel(r'$B_3$');
```
