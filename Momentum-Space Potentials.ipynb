```python
import numpy as np
import matplotlib.pyplot as plt

import helium4plusplus as he4
```

```python
R = he4.BETA6 / 2

s1 = he4.SemiLocalHelium4System(R, 0, 2, 3, 6)
```

```python
fig, ax = plt.subplots(figsize=(3, 3), dpi=250)
fig.patch.set_facecolor('white')

ax.imshow(s1.v_tilde);
```

```python
s2 = he4.NonlocalHelium4System(R, 0, 2, 3, 2)
```

```python
fig, ax = plt.subplots(figsize=(3, 3), dpi=250)
fig.patch.set_facecolor('white')

ax.imshow(s2.v_tilde);
```
