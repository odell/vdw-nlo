```python
import numpy as np
import matplotlib.pyplot as plt
import helium4plusplus as he4
```

```python
rgflow = np.loadtxt('datfiles/rg-flow-semilocal.dat')
```

```python
systems = [he4.SemiLocalHelium4System(R, 0, 2, 6, 6) for R in rgflow[:, 0]]
```

```python
spectra = [s.bound_state_spectrum(gi, gj) for (s, gi, gj) in zip(systems, rgflow[:, 1], rgflow[:, 2])]
```

```python
fig, ax = plt.subplots(dpi=300)
fig.patch.set_facecolor('white')

for (i, spec) in enumerate(spectra):
    for s in spec:
        ax.scatter(he4.BETA6/rgflow[i, 0], 1000*np.abs(s), color=f'C{i}')

ax.set_xlabel(r'$\beta_6/R$')
ax.set_ylabel(r'$B_2$ (mK)')
ax.set_yscale('log');
```
